```python
import pandas as pd
```

```python
data = pd.read_csv("Final Dataset.csv")
data.head()
```

```python
data.info()
```

```python
data = data[data.History_Of_Backlogs != " "]
```

```python
data = data.sort_values('History_Of_Backlogs')
data = data.sort_values("Work_Experience", ascending=False)
data = data.sort_values("10th_percentage", ascending=False)
data = data.sort_values("12th_percentage", ascending=False)
data = data.sort_values("Degree_Percentage", ascending=False)
data_final  = pd.DataFrame(data.head(25))
```

```python
data_final
```

```python
data_final["Gender"].value_counts().sort_index().plot(kind='pie', rot=0, ylabel='Gender',title = "10th Boards Vs Applicants")
```

```python
data_final["Degree_Field"].value_counts().sort_index().plot(kind='pie', rot=0, ylabel='Degree',title = "Degree analysis")
```

```python
data_final["Salary_expectations_per_month"].value_counts().sort_index().plot(kind='barh', rot=0, ylabel='Gender',title = "Salary Expectation of applicants", color = "black")
```
